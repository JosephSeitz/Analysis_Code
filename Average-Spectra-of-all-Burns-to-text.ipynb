```python
def speed_calculation(df):
    horizontal_speed = list(np.sqrt(df["V"]**2 + df["U"]**2)**2)
    return horizontal_speed

def kinectic_tke(df,x_fire, pre_fire):
    u_pre_lst, u_brn_lst, u_pst_lst = fire_period_split(list(df["U"]), x_fire, pre_fire)
    v_pre_lst, v_brn_lst, v_pst_lst = fire_period_split(list(df["V"]), x_fire, pre_fire)
    w_pre_lst, w_brn_lst, w_pst_lst = fire_period_split(list(df["W"]), x_fire, pre_fire)
    
    pre_tke_lst, brn_tke_lst, pst_tke_lst = np.full(len(u_pre_lst),np.nan), \
                    np.full(len(u_brn_lst),np.nan), np.full(len(u_pst_lst),np.nan)
    
    for i in range(len(u_pre_lst)):
        pre_tke_lst[i] = (u_pre_lst[i]**2 + v_pre_lst[i]**2 + w_pre_lst[i]**2)/2
        
    for i in range(len(u_brn_lst)):
        brn_tke_lst[i] = (u_brn_lst[i]**2 + v_brn_lst[i]**2 + w_brn_lst[i]**2)/2
    
    for i in range(len(u_pst_lst)):
        pst_tke_lst[i] = (u_pst_lst[i]**2 + v_pst_lst[i]**2 + w_pst_lst[i]**2)/2
    
    return pre_tke_lst, brn_tke_lst, pst_tke_lst

def horizontal_heat_flux(df, x_fire, pre_fire):
   
    w_pre_lst, w_brn_lst, w_pst_lst = fire_period_split(list(df["W"]), x_fire, pre_fire)
    t_pre_lst, t_brn_lst, t_pst_lst = fire_period_split(list(df["T"]), x_fire, pre_fire)
    
    pre_wt_lst, brn_wt_lst, pst_wt_lst = np.full(len(w_pre_lst),np.nan), \
        np.full(len(w_brn_lst),np.nan), np.full(len(w_pst_lst),np.nan)
    
    for i in range(len(w_pre_lst)):
        pre_wt_lst[i] = (w_pre_lst[i] * t_pre_lst[i])
        
    for i in range(len(w_brn_lst)):
        brn_wt_lst[i] = (w_brn_lst[i] * t_brn_lst[i])
        
    for i in range(len(w_pst_lst)):
        pst_wt_lst[i] = (w_pst_lst[i] * t_pst_lst[i])
        
    
    return pre_wt_lst, brn_wt_lst, pst_wt_lst
        
def horizontal_wind_mom(df, x_fire, pre_fire):
   
    w_pre_lst, w_brn_lst, w_pst_lst = fire_period_split(list(df["W"]), x_fire, pre_fire)
    s_pre_lst, s_brn_lst, s_pst_lst = fire_period_split(speed_calculation(df), x_fire, pre_fire)
    
    pre_ws_lst, brn_ws_lst, pst_ws_lst = np.full(len(w_pre_lst),np.nan), \
        np.full(len(w_brn_lst),np.nan), np.full(len(w_pst_lst),np.nan)

    for i in range(len(w_pre_lst)):
        pre_ws_lst[i] = (w_pre_lst[i] * s_pre_lst[i])
        
    for i in range(len(w_brn_lst)):
        brn_ws_lst[i] = (w_brn_lst[i] * s_brn_lst[i])
        
    for i in range(len(w_pst_lst)):
        pst_ws_lst[i] = (w_pst_lst[i] * s_pst_lst[i])
        
    
    return pre_ws_lst, brn_ws_lst, pst_ws_lst
        
def fire_period_split(var_lst, x_fire, pre_fire, period_t = 9000):
    
    t_start, f_end, t_end  =pre_fire-period_t, x_fire+period_t, (len(var_lst)-1)
    if t_start < 0:
        t_start = 0
    
    pre_lst = list(var_lst[int(t_start) : int(pre_fire)])
    brn_lst = list(var_lst[int(x_fire) : int(x_fire+period_t)])
    pst_lst = list(var_lst[int(x_fire+period_t) : int(x_fire+(period_t*2))])
    var_mean = np.nanmean(pre_lst)
        
    pre_lst -= var_mean
    brn_lst -= var_mean
    pst_lst -= var_mean
    
    return pre_lst, brn_lst, pst_lst
```

```python
def master_spectra(variable, sigma_fire, sig):
    path = "/Users/joeyp/Desktop/10X10_Truss_SERDP_Burns/JS_Cleaned_Data/Full-Timestamp/Burn-" #w10
    path_end = "/sonics/"
    sc_name_lst= ["A1","A2", "A3", "A4", "B1", "B2", "B3", "B4", \
                     "C1", "C2", "C3", "C4", "D1", "D2", "D3", "D4"]
    df_index_lst = []

    for brn in list(burn_d.keys()):
        for snc in sc_name_lst:
            df_index_lst.append(str("B"+brn+snc))

    df_pre_x_master, df_pre_y_master = pd.DataFrame(), pd.DataFrame()
    df_brn_x_master, df_brn_y_master = pd.DataFrame(), pd.DataFrame()
    df_pst_x_master, df_pst_y_master = pd.DataFrame(), pd.DataFrame()

    pre_snc_cnt, brn_snc_cnt, pst_snc_cnt = 0, 0, 0
    snc_cnt = 0
    for brn in range(len(list(burn_d.keys()))):
        burn = list(burn_d.keys())[brn]
        p = path+burn+ path_end
        files = text_finder(p)
        for i in range(16):

            df = file_to_df(p+files[i])

            if sigma_fire == True:
                
                if variable == "TKE":
                    pre_lst, brn_lst, pst_lst = kinectic_tke(df, burn_d[burn][1],burn_d[burn][0])
                if variable == "S'W'":
                    pre_lst, brn_lst, pst_lst = horizontal_wind_mom(df, burn_d[burn][1],burn_d[burn][0])
                if variable == "T'W'":
                    pre_lst, brn_lst, pst_lst = horizontal_heat_flux(df, burn_d[burn][1],burn_d[burn][0])
                if variable =="W" or variable == "U" or variable == "V" or variable == "T":
                    pre_lst, brn_lst, pst_lst = fire_period_split(list(df[variable]), burn_d[burn][1],burn_d[burn][0])
                if variable == "S":
                    pre_lst, brn_lst, pst_lst = fire_period_split(speed_calculation(df), burn_d[burn][1],burn_d[burn][0])

            if sigma_fire == False:
                x_fire = fire_start(df, burn_d[burn][0], sig = sig)
                if variable == "TKE":
                    pre_lst, brn_lst, pst_lst = kinectic_tke(df, x_fire, burn_d[burn][0])
                if variable == "S'W'":
                    pre_lst, brn_lst, pst_lst = horizontal_wind_mom(df, x_fire, burn_d[burn][0])
                if variable == "T'W'":
                    pre_lst, brn_lst, pst_lst = horizontal_heat_flux(df, x_fire, burn_d[burn][0])
                
                if variable =="W" or variable == "U" or variable == "V" or variable == "T":
                    pre_lst, brn_lst, pst_lst = fire_period_split(list(df[variable]), x_fire, burn_d[burn][0])
                if variable == "S":
                    pre_lst, brn_lst, pst_lst = fire_period_split(speed_calculation(df), x_fire, burn_d[burn][0])
            if variable =="W" or variable == "U" or variable == "V" or variable == "T":
                pre_x, pre_y = fast_fourier_spectra(np.multiply(pre_lst,pre_lst),redist=False)#binn)
                brn_x, brn_y = fast_fourier_spectra(np.multiply(brn_lst,brn_lst),redist=False)#binn)
                pst_x, pst_y = fast_fourier_spectra(np.multiply(pst_lst,pst_lst),redist=False)#binn)
            if variable =="TKE" or variable == "S'W'" or variable == "T'W'" or variable == "S":
                pre_x, pre_y = fast_fourier_spectra(pre_lst,redist=False)#binn)
                brn_x, brn_y = fast_fourier_spectra(brn_lst,redist=False)#binn)
                pst_x, pst_y = fast_fourier_spectra(pst_lst,redist=False)#binn)
            if len(pre_x) == 4500 and len(pre_y) ==4500:
                df_pre_x_master[df_index_lst[snc_cnt]+"_x"] = pre_x
                df_pre_y_master[df_index_lst[snc_cnt]+"_y"] = pre_y
                pre_snc_cnt+= 1
            if len(brn_x) == 4500 and len(brn_y) ==4500:
                df_brn_x_master[df_index_lst[snc_cnt]+"_x"] = brn_x
                df_brn_y_master[df_index_lst[snc_cnt]+"_y"] = brn_y
                brn_snc_cnt += 1
            if len(pst_x) == 4500 and len(pst_y) ==4500:
                df_pst_x_master[df_index_lst[snc_cnt]+"_x"] = pst_x
                df_pst_y_master[df_index_lst[snc_cnt]+"_y"] = pst_y
                pst_snc_cnt += 1

            snc_cnt+= 1
    print("PRE Sonics:",pre_snc_cnt, "BRN Sonics:", brn_snc_cnt,\
         "PST Sonics:", pst_snc_cnt)
            
    df_master = [df_pre_x_master, df_pre_y_master, df_brn_x_master, df_brn_y_master, \
                    df_pst_x_master, df_pst_y_master]
    return df_master

```

```python
def rebin_ave_master_spect(variable, sigma_fire, sig = 5, binn = 100, freq_function = True):
    df_master = master_spectra(variable, sigma_fire, sig)
    x1, y1 = df_master[0].mean(axis = 1), df_master[1].mean(axis = 1)
    x2, y2 = df_master[2].mean(axis = 1), df_master[3].mean(axis = 1)
    x3, y3 = df_master[4].mean(axis = 1), df_master[5].mean(axis = 1)

    new_x1, df_stat_1, new_y1 = lin_rb(x1,y1,9000,binn)
    new_x2, df_stat_2, new_y2 = lin_rb(x2,y2,9000,binn)
    new_x3, df_stat_3, new_y3 = lin_rb(x3,y3,9000,binn)

    if freq_function == True:
        spec_plt_lst =[[new_x1, new_y1*new_x1], [new_x2, new_y2*new_x2], [new_x3, new_y3*new_x3]]
    if freq_function == False:
        spec_plt_lst =[[new_x1, new_y1], [new_x2, new_y2], [new_x3, new_y3]]
    
    return spec_plt_lst 
```

```python
def data_for_plotting(variable, save = True):
    sig_val_lst,col_hdrs, all_sigs= [5,7.5,10,12.5,15], ["5_sig_First_SNC"],[]
    for i in range(len(sig_val_lst)):
        if i ==0:
            all_sigs.append(rebin_ave_master_spect(variable, sigma_fire=True))
            continue
        else:
            all_sigs.append(rebin_ave_master_spect(variable, sigma_fire=False, sig = sig_val_lst[i]))
            col_hdrs.append(str(sig_val_lst[i])+"_sig_Per_SNC")
    
    if save ==True:
        df, col_add = pd.DataFrame(), ["PRE","BRN", "PST"]
        for sg in range(len(all_sigs)):
            for prd in range(3):
                df[col_add[prd]+"_x_"+col_hdrs[sg]] = all_sigs[sg][prd][0]
                df[col_add[prd]+"_y_"+col_hdrs[sg]] = all_sigs[sg][prd][1]
        pre_cols, brn_cols, pst_cols =[], [], []
        for i in range(len(df.columns)):
            if "PRE" in str(df.columns[i]):
                pre_cols.append(df.columns[i])
            if "BRN"in str(df.columns[i]):
                brn_cols.append(df.columns[i])
            if "PST" in str(df.columns[i]):
                pst_cols.append(df.columns[i])
        df_new, cols = pd.DataFrame(), list(pre_cols+brn_cols+pst_cols)
        for i in cols:
            df_new[i] = df[i]
            
        df_new.to_csv(str(variable)+'_Averaged_Spectra_Varied_Sigmas.txt', index=None, sep=' ')
        
    return df_new
```
