```python
import matplotlib.pyplot as plt #used for creating plots
import matplotlib.patches as mpatches #used for plot legends
from commonly_used_functions import file_to_df
import numpy as np
%matplotlib inline
```

```python
def guide_ln_2_3(N = 10):
    x = np.logspace(-.5,.5,50)
    e_2_3= []
    for i in x:
        e_2_3.append((i*N**(-10))**((-2/3)))
    return x, e_2_3

def guide_ln_5_3(N = 10):    
    x, e_5_3 = np.logspace(-.5,.5,50)
    for i in x:
        e_5_3.append((i*(N*.1)**(10))**((-5/3)))
    return x, e_5_3

def guide_ln_7_3(N = 10):
    x = np.logspace(-.5,.5,50)
    e_7_3 = []
    for i in x:
        e_7_3.append((i*(N*.1)**(10))**((-7/3)))
    return x, e_7_3
```

```python
def guid_crct(variable):
    if variable == "TKE":
        return [2.8,2.8,3]
    if variable == "T":
        return [2.8,13,9]
    if variable =="T'W'":
        return [2.1,4.7,3.5]
    if variable == "S'W'":
        return [3.8,3.5,4]
    if variable == "S":
        return [3.6,3.8,3.7]
    if variable == "U" or variable == "V":
        return [4.8,5.4,5.9]
    if variable == "W":
        return [2.4,2.4,2.5]
```

```python
def plotting_spectra(variable, ncols = 1,norm = True, save = False):
    varience_lst = ["U_var","W_var","V_var"]
    p = "C:/Users/joeyp/Desktop/10X10_Truss_SERDP_Burns/Analysis/Spectral Analysis/"+ \
    str(variable)+ "_Averaged_Spectra_Varied_Sigmas.txt"
    
   # p = "C:/Users/joeyp/Desktop/10X10_Truss_SERDP_Burns/Analysis/Analysis_Code/"+\
   #         str(variable)+"_Averaged_Spectra_Varied_Sigmas.txt"
    df = file_to_df(p)

    sig_val_lst, sig_plt_label = [5,7.5,10,12.5,15],["5$\sigma$ First SNC"] 
    for i in range(1,len(sig_val_lst)):
        sig_plt_label.append(str(sig_val_lst[i])+"$\sigma$ Per SNC")
    lnstl = ["-","--","dotted","dashdot", (0, (3, 1, 1, 1, 1, 1))]    
    colors, fire_prd, guid_num = ["b", "r", "g"], ["PRE", "BRN", "PST"], guid_crct(variable)

    fig, axs = plt.subplots(nrows=1,ncols=ncols,figsize=(20,8),sharey = True)
    count, ymin, ymax = 0, 10**(10), 10**(-100)
    for j in range(3):
        if ncols ==1:
            ax= axs
        if ncols ==3:
            ax = axs[j]
        for i in range(0,10,2):
            y = df[df.columns[i+1+count]]
            if norm == True:
                print("Max:",max(y), "Norm", np.linalg.norm(y))
                y = y/np.linalg.norm(y)
            if np.min(y) < ymin:
                ymin = np.min(y)
            if np.max(y)> ymax:
                ymax = np.max(y)
            ax.loglog(df[df.columns[i+count]],\
                      y, c = colors[j],\
                      linestyle = lnstl[i//2],\
                       label = str(fire_prd[j]+" "+sig_plt_label[i//2]))

        guid_x, guid_y1= guide_ln_2_3(.65)
        ax.loglog(guid_x,guid_y1,c = "purple", label = "-2/3", linestyle = "--")
        ax.legend(loc = "lower left")
        ax.set_xlabel("Frequency (Hz)", fontsize = 16)
        if variable in varience_lst:
            ax.set_ylabel(r"$nS_{"+str(variable)[0]+"^2}(n)$", fontsize = 16)
        if variable not in varience_lst:
        
            ax.set_ylabel(r"$nS_{"+str(variable)+"}(n)$", fontsize = 16)
        ax.set_ylim(ymin, ymax)
        #ax.set_yscale("linear")
        ax.grid()
        count+=10
        
    if variable in varience_lst and norm ==True:
        fig.suptitle("Frequency Weighted Normalized Power Spectra of $" + str(variable)[0]+ "^2$", fontsize = 18)
    if variable not in varience_lst and norm ==True:
            fig.suptitle("Frequency Weighted Normalized Power Spectra of " + str(variable) , fontsize = 18)
    if variable in varience_lst and norm ==False:
        fig.suptitle("Frequency Weighted Power Spectra of $" + str(variable)[0]+ "^2$", fontsize = 18)
    if variable not in varience_lst and norm ==False:
            fig.suptitle("Frequency Weighted Power Spectra of " + str(variable) , fontsize = 18)
            
            
    fig.subplots_adjust(hspace=0.1, wspace=0.1)
    if save == True:
        if ncols == 1 and norm == False:
            fig.savefig("All_burn_spectra_"+str(variable)+"_one_axs.png",dpi=350)
        if ncols == 1 and norm == True:
            fig.savefig("Normalized_All_burn_spectra_"+str(variable)+"_one_axs.png",dpi=350)
        if ncols == 3 and norm == False:
            fig.savefig("All_burn_spectra_"+str(variable)+"_seperate_axs.png",dpi=350)
        if ncols == 3 and norm == True:
            fig.savefig("Normalized_All_burn_spectra_"+str(variable)+"_seperate_axs.png",dpi=350)
            
    plt.show()
```

```python
plotting_spectra("TKE")
plotting_spectra("T'W'")
plotting_spectra("S'W'")
plotting_spectra("T" )
plotting_spectra("S")
plotting_spectra("U_var")
plotting_spectra("V_var")
plotting_spectra("W_var")
```
